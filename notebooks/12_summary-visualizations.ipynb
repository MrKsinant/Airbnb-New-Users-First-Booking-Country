## Some visualizations

```python
# Represent nDCG mean score on training dataset:
fig, ax = plt.subplots(figsize=(8, 5))
x_values = np.arange(9)
width_rec = 0.5

# Configuration of the graph:
ax.set_title("nDCG mean score on training dataset")
ax.bar(x_values,
       ndcg_mean_train_list,
       width=width_rec,
       color='grey')
ax.set_xticks(x_values)
ax.set_xticklabels(labels_list, rotation=60)
ax.set_xlabel("List of classifiers")
ax.set_ylim(0, 1)
ax.yaxis.set_major_locator(mpl.ticker.MaxNLocator(11))
ax.yaxis.set_minor_locator(mpl.ticker.MaxNLocator(21))
ax.set_ylabel('Score')
ax.grid(color="grey", which="major", axis='y', linestyle='-', linewidth=0.25)
ax.grid(color="grey", which="minor", axis='y', linestyle=':', linewidth=0.25)

# Save the graph:
fig.savefig('../graphs/ndcg_mean_score_train.png', bbox_inches='tight')
```
